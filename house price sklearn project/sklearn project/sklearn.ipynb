```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x = [
    [7, 2],
    [8, 3],
    [10, 3],
    [11, 4]
]

y = [0,0,1,1]

model = DecisionTreeClassifier()

#Train the model
model.fit(x,y)

size = float(input("Enter the friut size"))
shade = float(input("Enter the color shade (1 - 10)"))

result = model.predict([[size,shade]])

if result == 0:
    print("Apple")
else:
    print("Orange")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

x = [1, 0, 1, 1, 0, 1, 0]

y = [1, 0, 1, 0, 0, 1, 1]

print("Accuracy = ",accuracy_score(x,y))
print("Pricision = ",precision_score(x,y))
print("Recall = ",recall_score(x,y))
print("F1 Score = ",f1_score(x,y))
```
